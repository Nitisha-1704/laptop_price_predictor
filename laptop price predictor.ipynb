```python
df['Memory'] = df['Memory'].astype(str).replace(r"\.0", "", regex=True)
df['Memory'] = df['Memory'].str.replace('GB', '', regex=False)
df['Memory'] = df['Memory'].str.replace('TB', '000', regex=False)

# Split and handle missing values
new = df['Memory'].str.split("+", n=1, expand=True)
df["first"] = new[0]
df["second"] = new[1].fillna("").astype(str)  # Fix for NoneType error

# Feature detection
df['hdd1'] = df["first"].apply(lambda x: 1 if "HDD" in x else 0)
df['ssd1'] = df["first"].apply(lambda x: 1 if "SSD" in x else 0)
df['hybrid1'] = df["first"].apply(lambda x: 1 if "Hybrid" in x else 0)
df['flash_storage1'] = df["first"].apply(lambda x: 1 if "Flash Storage" in x else 0)
df['first'] = df['first'].str.extract(r'(\d+)', expand=False).fillna('0')



df['hdd2'] = df["second"].apply(lambda x: 1 if "HDD" in x else 0)
df['ssd2'] = df["second"].apply(lambda x: 1 if "SSD" in x else 0)
df['hybrid2'] = df["second"].apply(lambda x: 1 if "Hybrid" in x else 0)
df['flash_storage2'] = df["second"].apply(lambda x: 1 if "Flash Storage" in x else 0)
df['second'] = df['second'].str.extract(r'(\d+)', expand=False).fillna('0')

df['HDD']=df["first"]*df['hdd1'] +df['second'] *df['hdd2']
df['SSD']=df["first"]*df['ssd1'] +df['second']*df['ssd2'] 
df['Hybrid']=df["first"]*df['hybrid1'] +df['second']*df['hybrid2'] 
df['Flash_Storage']=df["first"]*df['flash_storage1']+df['second']*df['flash_storage2']
df['second']=df['second'].str.replace(r'\D','')

df['first']=df['first'].astype(np.int32)
df['second']=df['second'].astype(np.int32)


df.drop(columns=['first','second','hdd1','ssd1','hybrid1','hybrid2','flash_storage1','flash_storage2','hdd2','ssd2'],inplace=True)

```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.15,random_state=2)
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import OneHotEncoder
from sklearn.metrics import r2_score,mean_absolute_error
from sklearn.linear_model import LinearRegression,Ridge,Lasso
from sklearn.neighbors import KNeighborsRegressor
```

```python
from sklearn.impute import SimpleImputer
```

```python
cat_cols = [0, 1, 7, 10, 11]
num_cols = [i for i in range(x_train.shape[1]) if i not in cat_cols]

# Step 1: Preprocessing
step1 = ColumnTransformer(transformers=[
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('encoder', OneHotEncoder(drop='first'))
    ]), cat_cols),
    
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='mean'))
    ]), num_cols)
], remainder='passthrough')

# Step 2: Model
step2 = LinearRegression()

# Pipeline
pipe = Pipeline([
    ('step1', step1),
    ('step2', step2)
])

# Fit and predict
pipe.fit(x_train, y_train)
y_pred = pipe.predict(x_test)

# Evaluation
print("R2 score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

```python
from sklearn.ensemble import VotingRegressor


# Define categorical and numerical columns
cat_cols = [0, 1, 7, 10, 11]
num_cols = [i for i in range(x_train.shape[1]) if i not in cat_cols]

# Shared preprocessing pipeline
preprocessor = ColumnTransformer(transformers=[
    ('cat', Pipeline([
        ('imputer', SimpleImputer(strategy='most_frequent')),
        ('onehot', OneHotEncoder(drop='first', handle_unknown='ignore'))
    ]), cat_cols),
    ('num', Pipeline([
        ('imputer', SimpleImputer(strategy='mean'))
    ]), num_cols)
])

# Define individual pipelines
pipe_lr = Pipeline([('preproc', preprocessor), ('model', LinearRegression())])
pipe_ridge = Pipeline([('preproc', preprocessor), ('model', Ridge(alpha=10))])
pipe_knn = Pipeline([('preproc', preprocessor), ('model', KNeighborsRegressor())])
pipe_dt = Pipeline([('preproc', preprocessor), ('model', DecisionTreeRegressor(random_state=42))])
pipe_rf = Pipeline([('preproc', preprocessor), ('model', RandomForestRegressor(random_state=42))])
pipe_gb = Pipeline([('preproc', preprocessor), ('model', GradientBoostingRegressor(random_state=42))])


# Create Voting Regressor
voting_regressor = VotingRegressor([
    ('lr', pipe_lr),
    ('ridge', pipe_ridge),
    ('knn', pipe_knn),
    ('dt', pipe_dt),
    ('rf', pipe_rf),
    ('gb', pipe_gb)
])

# Fit the voting regressor
voting_regressor.fit(x_train, y_train)

# Predict on test set
y_pred = voting_regressor.predict(x_test)

# Evaluate performance
print("R2 Score:", r2_score(y_test, y_pred))
print("MAE:", mean_absolute_error(y_test, y_pred))
```

```python
from tkinter import *
import joblib
import pandas as pd

def show_entry_fields():
    # Collect user input
    input_data = {
        'Company': e1.get(),
        'TypeName': e2.get(),
        'RAM': int(e3.get()),
        'Weight': float(e4.get()),
        'Touchscreen': e5.get(),
        'IPS': e6.get(),
        'PPI': float(e7.get()),
        'CPU Brand': e8.get(),
        'HDD': int(e9.get()),
        'SSD': int(e10.get()),
        'GPU Brand': e11.get(),
        'OS': e12.get()
    }

    # Convert to DataFrame
    df_input = pd.DataFrame([input_data])

    # Load trained pipeline
    pipe = joblib.load('pipe.pkl')

    # Predict price
    prediction = pipe.predict(df_input)[0]

    # Display result
    Label(master, text="Predicted Laptop Price (in ₹)").grid(row=14, column=0)
    Label(master, text=str(round(prediction, 2))).grid(row=14, column=1)
    print("Predicted Price:", prediction)

# Tkinter GUI setup
master = Tk()
master.title("Laptop Price Predictor")

Label(master, text="Laptop Price Predictor", bg="black", fg="white").grid(row=0, columnspan=2)

# Define feature labels
features = [
    "Company", "TypeName", "RAM (GB)", "Weight (kg)",
    "Touchscreen (Yes/No)", "IPS (Yes/No)", "PPI",
    "CPU Brand", "HDD (GB)", "SSD (GB)", "GPU Brand", "OS"
]

entries = []
for i, feature in enumerate(features):
    Label(master, text=feature).grid(row=i+1, column=0)
    entry = Entry(master)
    entry.grid(row=i+1, column=1)
    entries.append(entry)

# Assign entries to variables
e1, e2, e3, e4, e5, e6, e7, e8, e9, e10, e11, e12 = entries

Button(master, text='Predict', command=show_entry_fields).grid(row=13, columnspan=2)

mainloop()
```
